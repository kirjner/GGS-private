```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
from scipy import ndimage
import seaborn as sns
np.random.seed(42)
import os

n = 7
noise_scale = 0.4
alphabet = 'ABCDEFG'
noisy_reward_vec = np.eye(n) + (1 - np.eye(n))*np.random.normal(scale=noise_scale, size=(n,n))
```

```python
def true_noep_reward(x):
    return np.sum(x*np.eye(n), axis=(-1,-2))

def noisy_noep_reward(x):
    return np.sum(x * noisy_reward_vec, axis=(-1, -2))
```

```python
num_examples = 1000
seen_seqs = set()
data = []
n = 7
while len(data) < num_examples:
    oh_seq = np.eye(n)
    oh_seq = np.apply_along_axis(np.random.permutation, 1, oh_seq)
    oh_seq_str = ''.join(map(str, oh_seq))
    #sum of diagonal elements can be at most 2
    if np.sum(np.diag(oh_seq)) > 1 or oh_seq_str in seen_seqs:
        continue
    #convert to 2d array, then to string sequence based on alphabet above
    seen_seqs.add(oh_seq_str)
    seq_tokens = np.argmax(oh_seq, axis = 1)
    seq = ''.join([alphabet[i] for i in seq_tokens])
    seq_reward = noisy_noep_reward(oh_seq)
    data.append((seq, seq_reward))
```

```python
landscape_noep_constrained = pd.DataFrame(data, columns = ['sequence', 'target'])
landscape_noep_constrained['augmented'] = 0
landscape_noep_constrained.to_csv('/home/gridsan/akirjner/protein_design/GGS-private/data/Diagonal/landscape_noep_constrained.csv', index=False)
```

## No Epistasis, Constrained Starting Data

```python

K = 128
seeds = np.arange(1,6)
ROOT_DIR = '/home/gridsan/akirjner/protein_design/GGS-private/ckpt/Diagonal/mutations_0/percentile_0.0_1.0/unsmoothed/no_epistasis_constrained/samples/run_2023-08-07_20-40'

```

```python
seeds = np.arange(1,6)
def get_scores(rootdir: str, K: int):
    all_true_scores = np.zeros((5, 15))
    all_noisy_scores = np.zeros((5, 15))
    for i, seed in enumerate(seeds):
        seed_df = pd.read_csv(os.path.join(rootdir, f'seed_{seed}.csv'))
        epochs = seed_df['epoch'].unique()
        
        for epoch in epochs:
            epoch_df = seed_df[seed_df['epoch'] == epoch]
            topK_onehot = []
            seen_seqs = set()
            
            for seq in epoch_df.sort_values(by=['mutant_scores'], ascending=False)['mutant_sequences'].tolist():
                if seq in seen_seqs:
                    continue
                seen_seqs.add(seq)
                topK_onehot.append(np.array([np.array([1 if char == letter else 0 for char in alphabet]) for letter in seq]))
                if len(topK_onehot) == K:
                    break
                    
            topK_onehot = np.array(topK_onehot)
            true_scores = true_noep_reward(topK_onehot)
            noisy_scores = noisy_noep_reward(topK_onehot)
            
            all_true_scores[i, int(epoch)-1] = np.mean(true_scores)
            all_noisy_scores[i, int(epoch)-1] = np.mean(noisy_scores)
            
    return all_true_scores, all_noisy_scores

# Example usage:
# You would need to define the required parameters (seeds, true_noep_reward, noisy_noep_reward, etc.)
# true
```

```python
all_unsmoothed_true_scores, all_unsmoothed_noisy_scores = get_scores(ROOT_DIR, 128)
```

```python
SMOOTHED_ROOT_DIR = '/home/gridsan/akirjner/protein_design/GGS-private/ckpt/Diagonal/mutations_0/percentile_0.0_1.0/smoothed/no_epistasis_constrained/samples/run_2023-08-07_21-06'
```

```python
all_smoothed_true_scores, all_smoothed_noisy_scores = get_scores(SMOOTHED_ROOT_DIR, 128)
```

```python
avg_unsmoothed_true_scores = np.mean(all_unsmoothed_true_scores, axis = 0)
avg_unsmoothed_noisy_scores = np.mean(all_unsmoothed_noisy_scores, axis = 0)
avg_smoothed_true_scores = np.mean(all_smoothed_true_scores, axis = 0)
avg_smoothed_noisy_scores = np.mean(all_smoothed_noisy_scores, axis = 0)
```

```python
fig, axs  = plt.subplots(1, 2, figsize = (20, 10))
x = np.arange(1, 16)
axs[0].plot(x, avg_unsmoothed_true_scores, label = 'unsmoothed')
axs[0].plot(x, avg_smoothed_true_scores, label = 'smoothed')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('True Fitness (Avg. over 5 seeds)')
axs[0].set_title('True Fitness of top 128 sequences vs. Epoch')
axs[0].set_xticks(x)
axs[0].legend()

axs[1].plot(x, avg_unsmoothed_noisy_scores, label = 'unsmoothed')
axs[1].plot(x, avg_smoothed_noisy_scores, label = 'smoothed')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Noisy Fitness (Avg. over 5 seeds)')
axs[1].set_title('Noisy Fitness of top 128 sequences vs. Epoch')
axs[1].set_xticks(x)
axs[1].legend()
plt.suptitle("Constrained Non-Epistatic Landscape", fontsize = 36)
```
